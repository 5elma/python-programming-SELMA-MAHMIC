```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
f = np.sin(x)
f
```

```python
plt.figure(dpi = 100)
plt.plot(x, f)
plt.title("sin(x)")
plt.xlabel("x")
plt.ylabel("f")
plt.show()
```

```python
fig, ax = plt.figure(dpi=100), plt.axes()
ax.plot(x, f)
ax.set(xlim = (-6, 6), ylim = (-1.5, 1.5), xlabel="x", ylabel="f", title="sin(x)")
plt.show()
```

```python
plt.figure(figsize=(16,5), dpi=100)
plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.title("Plots")
plt.legend(("sin(x)", "cos(x)"))

plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
plt.show()
```

```python
fig, ax = plt.subplots(1,2, dpi=150, figsize=(10, 4))
rng = np.random.RandomState(seed=42)
data_normal_dist = rng.randn(1000)

ax[0].hist(data_normal_dist, bins=30, alpha=0.5)
ax[0].set(title="Histogram", xlabel="x", ylabel="y")

size = 1000*rng.rand(100)

sc = ax[1].scatter(x=rng.randn(100), y=rng.randn(100), c = rng.randn(100), alpha = 0.3, cmap="viridis", s=size)
fig.colorbar(sc)
ax[1].set(title="Bubble chart", xlabel="x", ylabel="y")

plt.savefig("../assets/Beautiful_plots.png", transparent=True, facecolor="white")
```

```python
class Student:
    def __init__(self, name):
        self.name = name

    def study(self):
        return f"{self.name} is studying"
    
print(Student.study)

s1 = Student("Ada")
print(s1.name)
print(s1.study)

print(s1.study())

print(Student.study(s1))
```
